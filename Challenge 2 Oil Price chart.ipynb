```python
import requests
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import pylab
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import numpy as np
%matplotlib inline

import json

with open('outputfile.json') as f:
  data = json.load(f)

print(data)

dates = [item['created_at'] for item in data['data']['prices']]
values = [item['price'] for item in data['data']['prices']]

rng = pd.date_range('2020-05-05T00:00:00.000Z', periods=30, freq='d')
ax = pd.DataFrame(data = np.random.randn(len(rng)), index=pd.to_datetime(rng))

df = pd.DataFrame(({'dates':dates, 'values':values}))
df['dates']  = [pd.to_datetime(i) for i in df['dates']]
df['dates'] = pd.to_datetime(df['dates'], unit='s')
df['mdates'] = [mdates.date2num(d) for d in df['dates']]
df.info()
fig, ax = plt.subplots()
ax.xaxis.set_major_formatter(DateFormatter('%d/%m'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation=90)
plt.plot(dates, values)
plt.gcf().set_size_inches(10, 5)
plt.xlabel('Time')
plt.ylabel('Price')
plt.title('Daily Average Oil Price')
```
